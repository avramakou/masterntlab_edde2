```
# Define stopwords
gr_stop_words = nltk.corpus.stopwords.words('greek')
gr_stop_words = gr_stop_words + ['amp', 'co', 'σας', 'αυτές', 'μας', 'γιατί', 'στις', 'είναι', 'από', 'ότι', 'και', 'το', 'τα', 'στο', 'του', 'της', 'για', 'στην', 'τους', 'αλλά', 'μια', 
                                 'ένα', 'τη', 'μας', 'αυτό', 'στον', 'έχει', 'ήταν', 'αυτή', 'στα', 'ας', 'πιο', 'αυτές', 'σε', 'τις', 'οι', 'το', 'τέσσερις', 'δείτε']
```

```
# Turn column into list
textColumn = '.'.join(final_mediaGR_df.title.tolist())
print(textColumn)
```

```
# Strip accents
import unicodedata
def strip_accents(s):
   return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
```

```
# Create list without accents
all_text_no = strip_accents(textColumn)
```

```
# Split list's words
wordList = re.sub("[^\w]", " ", all_text_no).split()
print(wordList)
```

```
# Remove stopwords from list
tokens_without_sw = [word for word in wordList if not word in gr_stop_words]
print(tokens_without_sw)
```

```
# Convert list's elements into lowercase
for i in range(len(tokens_without_sw)):
  tokens_without_sw[i] = tokens_without_sw[i].lower()
print(tokens_without_sw)
```

```
# Count 15 most common words (article titles)
c = Counter(tokens_without_sw)
most_common = c.most_common(20)
print(most_common)
```

```
# Create wordcloud (article titles)
unique_string=(" ").join(tokens_without_sw)
wordcloud = WordCloud(width = 1000, height = 500, collocations=False, background_color="black", stopwords = gr_stop_words).generate(unique_string)
plt.figure(figsize=(20,10)), 
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("your_file_name"+".png", bbox_inches='tight')
plt.show()
plt.close() # Remember: in order to not detect collocations -> collocations=False
```

```
# Create wordcloud (article titles) # Collocations are visible
unique_string=(" ").join(tokens_without_sw)
wordcloud = WordCloud(width = 1000, height = 500, collocations=True, background_color="white", stopwords = gr_stop_words).generate(unique_string) 
plt.figure(figsize=(20,10)), 
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("your_file_name"+".png", bbox_inches='tight')
plt.show()
plt.close() 
```
